```python
# List of categorical columns to visualize
categorical_columns = [
    'Reached_3w', 'Reached_3m',
    'Has_Payroll', 'Has_Investment', 'Has_Visa', 'Has_mortgage', 'Has_Deposit'
]

# Function to plot the distribution of categorical columns with adjusted annotations
def plot_categorical_columns_with_adjusted_annotations(data, title):
    nrows = (len(categorical_columns) + 1) // 2  # Arrange the plots in 2 columns
    fig, axes = plt.subplots(nrows, 2, figsize=(14, (nrows + 1) * 4))  # Increase scale
    fig.suptitle(title, fontsize=16)
    axes = axes.flatten()  # Flatten the 2D array of axes to 1D

    # Iterate over each categorical column and plot the distribution with annotations
    for i, col in enumerate(categorical_columns):
        ax = sns.countplot(x=col, data=data, ax=axes[i], palette='viridis')
        axes[i].set_title(f'Distribution of {col}')
        axes[i].set_ylabel('Count')

        # Adding annotations
        for p in ax.patches:
            ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center', fontsize=11, color='black', xytext=(0, 5),
                        textcoords='offset points')

    # Remove the last sub-plot if the total number of categorical columns is odd
    if len(categorical_columns) % 2 != 0:
        fig.delaxes(axes[-1])

    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust the spacing
    plt.show()

# Plot the distribution of categorical columns with adjusted annotations for both datasets
plot_categorical_columns_with_adjusted_annotations(data_3m, 'Distribution of Categorical Variables in 3-Months Dataset')
plot_categorical_columns_with_adjusted_annotations(data_6m, 'Distribution of Categorical Variables in 6-Months Dataset')
```

The categorical variable distributions in the 3-months and 6-months datasets show the following patterns:

**3-Months Dataset:**

* Reached_3w and Reached_3m: A majority of customers were not reached at both the 3-week and 3-month marks.

* Has_Payroll: The number of customers having payroll is slightly higher than those who don't.

* Has_Investment and Has_mortgage: A majority of customers do not have investments or mortgages.

* Has_Visa and Has_Deposit: Most customers have a deposit, and around half of the customers have a Visa.

**6-Months Dataset:**

* The distributions of categorical variables in the 6-months dataset are similar to those in the 3-months dataset, with minor variations.

```python
# List of numerical columns to visualize
numerical_columns = [
    'FUA_Balance', 'Number_Of_Services', 'Income',
    'VISA_balance', 'Number_Transactions', 'Age'
]

# Function to plot the distribution of numerical columns
def plot_numerical_columns(data, title):
    nrows = (len(numerical_columns) + 1) // 2  # Arrange the plots in 2 columns
    fig, axes = plt.subplots(nrows, 2, figsize=(14, nrows * 4))
    fig.suptitle(title, fontsize=16)
    axes = axes.flatten()  # Flatten the 2D array of axes to 1D

    # Iterate over each numerical column and plot the distribution
    for i, col in enumerate(numerical_columns):
        if col in data.columns:  # Check if the column exists in the dataset
            sns.histplot(data[col], bins=30, kde=True, ax=axes[i], color='teal')
            axes[i].set_title(f'Distribution of {col}')
            axes[i].set_xlabel(col)
            axes[i].set_ylabel('Frequency')

    # Remove the last sub-plot if the total number of numerical columns is odd
    if len(numerical_columns) % 2 != 0:
        fig.delaxes(axes[-1])

    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust the spacing
    plt.show()

# Plot the distribution of numerical columns for both datasets
plot_numerical_columns(data_3m, 'Distribution of Numerical Variables in 3-Months Dataset')
plot_numerical_columns(data_6m, 'Distribution of Numerical Variables in 6-Months Dataset')
```

Here are the distributions of the numerical variables in both datasets:

**3-Months Dataset:**


* FUA_Balance: Most of the customers have a lower Funds Under Administration (FUA) balance, indicating a right-skewed distribution.

* Number_Of_Services: Most customers have 1 or 2 services, with few having more than 2.

* Income: The distribution of income is right-skewed, with most customers having lower incomes and a few having higher incomes.

* VISA_balance: The distribution is right-skewed, with a majority of customers having lower Visa balances.

* Number_Transactions: This variable is also right-skewed, indicating that most customers have a lower number of transactions.

* Age: The age distribution is relatively normal, with a peak around the middle-age range.

**6-Months Dataset:**

* FUA_Balance, Number_Of_Services, VISA_balance, Number_Transactions, and Age: The distributions in the 6-months dataset are similar to those in the 3-months dataset.

*  Income: This variable is not available in the 6-months dataset.





```python
#A correlation matrix will help us understand the linear relationship between different variables.
#A positive value indicates a positive linear relationship, a negative value indicates a negative linear relationship, and a value close to zero indicates no linear relationship.

# Function to plot the correlation matrix of numerical columns
def plot_correlation_matrix(data, title):
    # Calculate the correlation matrix
    corr_matrix = data[numerical_columns].corr()

    # Plot the heatmap of the correlation matrix
    plt.figure(figsize=(10, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title(title)
    plt.show()

# Plot the correlation matrix for both datasets
plot_correlation_matrix(data_3m, 'Correlation Matrix of Numerical Variables in 3-Months Dataset')
plot_correlation_matrix(data_6m, 'Correlation Matrix of Numerical Variables in 6-Months Dataset')
```

Here are the correlation matrices for the numerical variables in both datasets:

**3-Months Dataset:**

* FUA_Balance has a moderate positive correlation with Number_Of_Services (0.32) and a weak positive correlation with VISA_balance (0.26) and Number_Transactions (0.21).

* Number_Of_Services has a weak positive correlation with Number_Transactions (0.22).

* Income has a very weak positive correlation with Number_Transactions (0.12) and Age (0.10).

**6-Months Dataset:**

* Similar correlations are observed in the 6-months dataset, with FUA_Balance having a moderate positive correlation with Number_Of_Services (0.33) and a weak positive correlation with VISA_balance (0.29) and Number_Transactions (0.21).

**Insights:**

* Customers with higher FUA_Balance tend to have more services and higher VISA_balance and conduct more transactions.

* Customers with more services tend to conduct more transactions.
* Income, where available, doesn’t show a strong correlation with other variables.
